# **What is the profit(average value, not real data from  steam) for each Tag and analysis for that?**

```python
tags = games_data['tag'].str.get_dummies(';')
```

```python
games_data['positive_pct'] = (
games_data['total_positive'] /
games_data['total_reviews']
) * 100
```

```python
games_data['negative_pct'] = (
games_data['total_negative'] /
games_data['total_reviews']
) * 100
```

```python
results = []

for tag_name in tags.columns:

    mask = (tags[tag_name] == 1)
    tagged_games = games_data.loc[mask]


    if tagged_games.empty:
        continue


    results.append({
        'tag': tag_name,
        'positive%': tagged_games['positive_pct'].mean().round(2),
        'negative%': tagged_games['negative_pct'].mean().round(2),
        'avg_price': tagged_games['price'].mean().round(2),
        'Total Reviews': tagged_games['total_reviews'].sum(),
        'Game Count': len(tagged_games)
    })


tag_df = pd.DataFrame(results)


tag_df = tag_df.sort_values('Total Reviews', ascending=False).reset_index(drop=True)
```

```python
tag_df['Average Profit (M)']=((tag_df['avg_price']*tag_df['Total Reviews']*0.4)/1000000).round(2)  ## Assume profit is 40% of total price
tag_df['Average Profit per game (M)']=(tag_df['Average Profit (M)']/tag_df['Game Count']).round(2)
```

```python
tag_df
```

```python
sorted_df = tag_df.sort_values('Average Profit (M)', ascending=False)

# Create a figure with subplots
fig4 = make_subplots(
    rows=2, cols=2,
    specs=[[{"type": "bar"}, {"type": "scatter"}],
           [{"type": "bar", "colspan": 2}, None]],
    subplot_titles=("Average Profit by Tag (Millions $)",
                   "Profit vs. Review Sentiment",
                   "Average Profit per Game by Tag (Millions $)"),
    vertical_spacing=0.12,
    horizontal_spacing=0.08
)

# 1. Bar chart for Average Profit by Tag
fig4.add_trace(
    go.Bar(
        x=sorted_df['tag'],
        y=sorted_df['Average Profit (M)'],
        marker_color=sorted_df['positive%'],
        marker_colorbar=dict(title="Positive %"),
        marker_colorscale='RdYlGn',
        text=sorted_df['Average Profit (M)'].apply(lambda x: f"${x:.1f}M"),
        textposition='auto',
        name='Total Profit'
    ),
    row=1, col=1
)

# 2. Scatter plot for Profit vs Sentiment with size as Total Reviews
fig4.add_trace(
    go.Scatter(
        x=sorted_df['positive%'],
        y=sorted_df['Average Profit (M)'],
        mode='markers',
        marker=dict(
            size=sorted_df['Total Reviews'].apply(lambda x: min(max(x/10000, 10), 50)),
            color=sorted_df['avg_price'],
            colorscale='Viridis',
            colorbar=dict(title="Avg Price ($)"),
            showscale=True
        ),
        text=sorted_df['tag'],
        hovertemplate='<b>%{text}</b><br>Positive: %{x:.1f}%<br>Profit: $%{y:.1f}M<br>Reviews: %{marker.size}<br>Avg Price: $%{marker.color:.2f}<extra></extra>',
        name='Tag Performance'
    ),
    row=1, col=2
)

# 3. Bar chart for Average Profit per Game
fig4.add_trace(
    go.Bar(
        x=sorted_df['tag'],
        y=sorted_df['Average Profit per game (M)'],
        marker_color=sorted_df['Game Count'],
        marker_colorscale='Blues',
        marker_colorbar=dict(title="Game Count"),
        text=sorted_df['Average Profit per game (M)'].apply(lambda x: f"${x:.2f}M"),
        textposition='auto',
        name='Profit per Game'
    ),
    row=2, col=1
)

# Update layout
fig4.update_layout(
    title_text="Game Tag Performance Analysis",
    height=900,
    width=1200,
    template="plotly_white",
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    font=dict(family="Arial", size=12)
)

# Update axes
fig4.update_xaxes(tickangle=45, row=1, col=1)
fig4.update_xaxes(tickangle=45, row=2, col=1)
fig4.update_xaxes(title_text="Positive Reviews (%)", row=1, col=2)
fig4.update_yaxes(title_text="Average Profit (Millions $)", row=1, col=1)
fig4.update_yaxes(title_text="Average Profit (Millions $)", row=1, col=2)
fig4.update_yaxes(title_text="Average Profit per Game (Millions $)", row=2, col=1)

# Show the figure
fig4.show()
```

# **ML model (Predict tag)**

```python
merged_data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
X=merged_data[['recommendation','review_score','total_playtime_hours','total_positive','total_negative']]
y=merged_data['tag']
```

```python
KN=KNeighborsClassifier(n_neighbors=11, p=2, metric='euclidean')
NB=GaussianNB()
RF=RandomForestClassifier()
DT = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
x_train2,x_test2,y_train2,y_test2=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
x_train3,x_test3,y_train3,y_test3=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
x_train4,x_test4,y_train4,y_test4=train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
KN.fit (x_train,y_train)
NB.fit   (x_train2,y_train2)
RF.fit  (x_train3,y_train3)
DT.fit  (x_train4,y_train4)
```

KNN

```python
y1=KN.predict(x_test)
score=round(accuracy_score(y1,y_test), 4)
print(score)
```

GaussianNB

```python
y2=NB.predict(x_test2)
score2=round(accuracy_score(y2,y_test2),4)
print(score2)
```

RandomForestClassifier

```python
y3=RF.predict(x_test3)
score3=round(accuracy_score(y3,y_test3),4)
print(score3)
```

DecisionTreeClassifier

```python
y4=DT.predict(x_test4)
score4=round(accuracy_score(y4,y_test4),4)
print(score4)
```

```python
model_names = ['KNN', 'GaussianNB', 'RandomForestClassifier', 'DecisionTreeClassifier']
accuracies = [score, score2, score3, score4]

accuracy_df = pd.DataFrame({
    'Model': model_names,
    'Accuracy': accuracies
})

display(accuracy_df)
```

```python
fig = px.bar(
    accuracy_df.sort_values(by='Accuracy', ascending=True), # Sort ascending for horizontal bar chart display
    x='Accuracy',
    y='Model',
    orientation='h',
    title='Model Accuracy Comparison (Horizontal Bar Chart)',
    labels={'Model': 'Machine Learning Model', 'Accuracy': 'Accuracy Score'},
    color='Accuracy',
    color_continuous_scale='twilight',
    height=600
)



fig.update_layout(
    xaxis_title='Accuracy Score',
    yaxis_title='Machine Learning Model',
    xaxis_range=[0, 1.05],
    font=dict(size=12)
)

fig.update_traces(texttemplate='%{x:.4f}', textposition='outside')

fig.show()
```

```python
aaa_df = tag_df[tag_df['tag'].str.contains('AAA_')].sort_values('Average Profit (M)', ascending=False)

fig_aaa = make_subplots(
    rows=2, cols=2,
    specs=[[{"type": "bar"}, {"type": "scatter"}],
           [{"type": "bar", "colspan": 2}, None]],
    subplot_titles=("AAA - Average Profit by Tag (Millions $)",
                   "AAA - Profit vs. Review Sentiment",
                   "AAA - Average Profit per Game by Tag (Millions $)"),
    vertical_spacing=0.12,
    horizontal_spacing=0.08
)

# 1. Bar chart for Average Profit by Tag (AAA)
fig_aaa.add_trace(
    go.Bar(
        x=aaa_df['tag'],
        y=aaa_df['Average Profit (M)'],
        marker_color=aaa_df['positive%'],
        marker_colorbar=dict(title="Positive %"),
        marker_colorscale='RdYlGn',
        text=aaa_df['Average Profit (M)'].apply(lambda x: f"${x:.1f}M"),
        textposition='auto',
        name='Total Profit'
    ),
    row=1, col=1
)

# 2. Scatter plot for Profit vs Sentiment with size as Total Reviews (AAA)
fig_aaa.add_trace(
    go.Scatter(
        x=aaa_df['positive%'],
        y=aaa_df['Average Profit (M)'],
        mode='markers',
        marker=dict(
            size=aaa_df['Total Reviews'].apply(lambda x: min(max(x/10000, 10), 50)),
            color=aaa_df['avg_price'],
            colorscale='Viridis',
            colorbar=dict(title="Avg Price ($)"),
            showscale=True
        ),
        text=aaa_df['tag'],
        hovertemplate='<b>%{text}</b><br>Positive: %{x:.1f}%<br>Profit: $%{y:.1f}M<br>Reviews: %{marker.size}<br>Avg Price: $%{marker.color:.2f}<extra></extra>',
        name='Tag Performance'
    ),
    row=1, col=2
)

# 3. Bar chart for Average Profit per Game (AAA)
fig_aaa.add_trace(
    go.Bar(
        x=aaa_df['tag'],
        y=aaa_df['Average Profit per game (M)'],
        marker_color=aaa_df['Game Count'],
        marker_colorscale='Blues',
        marker_colorbar=dict(title="Game Count"),
        text=aaa_df['Average Profit per game (M)'].apply(lambda x: f"${x:.2f}M"),
        textposition='auto',
        name='Profit per Game'
    ),
    row=2, col=1
)

# Update layout
fig_aaa.update_layout(
    title_text="AAA Game Tag Performance Analysis",
    height=900,
    width=1200,
    template="plotly_white",
    legend=dict(orientation="h", yanchor="bottom", y=1.02, xanchor="right", x=1),
    font=dict(family="Arial", size=12)
)

# Update axes
fig_aaa.update_xaxes(tickangle=45, row=1, col=1)
fig_aaa.update_xaxes(tickangle=45, row=2, col=1)
fig_aaa.update_xaxes(title_text="Positive Reviews (%)", row=1, col=2)
fig_aaa.update_yaxes(title_text="Average Profit (Millions $)", row=1, col=1)
fig_aaa.update_yaxes(title_text="Average Profit (Millions $)", row=1, col=2)
fig_aaa.update_yaxes(title_text="Average Profit per Game (Millions $)", row=2, col=1)

fig_aaa.show()
```
